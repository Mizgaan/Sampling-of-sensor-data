```python
def get_file_list(output_dir, prefix="", suffix="", file_basename="feature_data", file_format="xls"): #  =file_basename =feature_data, file_format="xls"

    dir_list=[]
    # print('walk_dir',os.walk(output_dir))
    for path,dirs,files in os.walk(output_dir):
        for d in dirs:
            full_path=os.path.abspath(os.path.join(path,d))
            dir_list.append(full_path)
            
    #dir_list = [os.path.join(output_dir, r"%s2CARS%s"        % (prefix,suffix)),
     #           os.path.join(output_dir, r"%s3CARS%s"        % (prefix,suffix)),
     #           os.path.join(output_dir, r"%sCITY%s"         % (prefix,suffix)),
     #           os.path.join(output_dir, r"%sCITY_V2%s"      % (prefix,suffix)),
     #           os.path.join(output_dir, r"%sOVERHEAD_V2%s"  % (prefix,suffix)),
     #           os.path.join(output_dir, r"%sARS410_MIXED%s" % (prefix,suffix))]
    
    #print suffix     
    print ('dir_list:- ',(dir_list))
    files = []
    for dir in dir_list:
        if(os.path.isdir(dir)):
            filenames = [f for f in os.listdir(dir) if os.path.isfile(os.path.join(dir, f))]
            for f in filenames:
                if (file_format == os.path.splitext(f)[1][1:]):
                    if (file_basename in f):
                        files.append(os.path.join(dir, f))
                        print("--> %s" % files[-1])
    return files
```

## Downsampling

```python
downsample_data=pd.DataFrame()
for c in classes:
    
    if not c==min_class:
        a=int(Prob_class_distribution[c]/Prob_class_distribution[min_class]* Min_data)
#         n=temp[c]-a
        print('C-->',c,':',a)
        sample_data=resample(x[x['class']==c],n_samples=a)
        downsample_data=pd.concat([downsample_data,sample_data],axis=0)
        
        
    else:
        downsample_data=pd.concat([downsample_data,x[x['class']==c]],axis=0)
```

```python
acc_score=[]
for train_index , test_index in kf_val.split(X):
    X_train , X_test = X.iloc[train_index,:],X.iloc[test_index,:]
    y_train , y_test = y.iloc[train_index] , y.iloc[test_index]
     
    model.fit(X_train,y_train)
    pickle.dump((model), open('D:\Decision_TREE_PC.pkl', 'wb'))    
    pred_values = model.predict(X_test)
     
    acc = accuracy_score(pred_values , y_test)  #model.score(X_test,y_test)
    acc_score.append(acc)
    
    rep=classification_report(pred_values , y_test)
    print(rep)
    print(f'No.of Data Points Of train:-{len(train_index)} & No.of Data Points Of test:-{len(test_index)}')
        
```
